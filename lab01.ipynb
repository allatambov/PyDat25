В файле `languages.csv` сохранены данные по вымирающим/вымершим языкам и языкам на грани исчезновения. Одна строка таблицы соответствует одному языку. 

Показатели в файле:

* `name_en`: название языка на английском;
* `countries`: страны, на которых говорят на данном языке;
* `status`: статус языка – степень угрозы существованию языка согласно ЮНЕСКО, см. ниже;
* `speakers`: число говорящих на языке;
* `location`: описание ареала языка (где распространен);
* `lat` и `lon` : широта и долгота для `location`.

Пояснения для значений в `status`:

* 1: `Vulnerable`, большинство детей говорят на языке, но чаще всего в ограниченных сферах деятельности (например, дома); 
* 2: `Definitely endangered`, дети больше не изучают язык как родной дома; 
* 3: `Severely endangered`, на языке говорит старшее поколение (бабушки/дедушки), родители могут язык понимать, но не говорить на нем со своими детьми; 
* 4: `Critically endangered`, самые молодые носители языка - старшее поколение (бабушки/дедушки и старше), но они говорят на языке редко;
* 5: `Extinct`, говорящих на языке не осталось.

Ниже написан код, который загружает данные из файла по ссылке, забирает из датафрейма столбцы с названиями `status` и `speakers`, а из столбцов – массивы значений.

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/allatambov/PyDat25/refs/heads/main/languages.csv")

speakers = df["speakers"].values
status = df["status"].values

print(speakers)
print(status)
```
